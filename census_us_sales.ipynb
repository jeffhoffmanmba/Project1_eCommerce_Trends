```python
import pandas as pd
import numpy as np
import datetime
```

```python
us_data = "Resource/final_us_data.csv"
```

```python
us_data_df = pd.read_csv(us_data)
```

```python
us_data_df.head()
```

```python
reduced_us_df = us_data_df[["per_name", "cat_code", "cat_desc", "dt_desc","dt_unit", "geo_code", "val", "adj_desc"]]
reduced_us_df.head()
```

```python
renamed_us_df = reduced_us_df.rename(columns={"per_name": "Month/Year",
                                            "cat_code": "Category Code",
                                            "cat_desc": "Category Description",
                                            "dt_desc": "Data Type",
                                            "dt_unit": "Units",
                                            "geo_code": "Country",
                                            "val": "Total Sales",
                                             "adj_desc": "Adjusted Description"})
renamed_us_df.head()
```

```python
electronic_shopping_df = renamed_us_df.loc[renamed_us_df["Category Code"] == "4541"]
electronic_shopping_df.head()
```

```python
mon_sales_df = electronic_shopping_df.loc[electronic_shopping_df["Data Type"] == "Sales - Monthly"]
mon_sales_df.tail()
```

```python
adj_desc_df = mon_sales_df.loc[mon_sales_df["Adjusted Description"] == "not adjusted"]
adj_desc_df.head()
```

```python
adj_desc_df[["Month", "Date", "Year"]]= adj_desc_df["Month/Year"].str.split("/", expand=True)
adj_desc_df.head(5)
```

```python
tenyr_us_df = adj_desc_df.loc[adj_desc_df["Year"] >= "2010"]
tenyr_us_df.head()
```

```python
tenyr_us_df["Month"].replace({"1":"January",
                              "2":"February",
                              "3":"March",
                              "4":"April",
                              "5":"May",
                              "6":"June",
                              "7":"July",
                              "8":"August",
                              "9":"September",
                              "10":"October",
                              "11":"November",
                              "12":"December"}, inplace=True)
tenyr_us_df.head()
```

```python
clean_us_df = tenyr_us_df[["Month","Year","Category Code","Category Description","Country","Total Sales","Units"]]
clean_us_df.head(12)
```
